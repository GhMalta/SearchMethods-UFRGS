**Metodos de Busca Cega**

1) Busca em Largura



```python
from collections import defaultdict
import numpy as np

class fila_circular():
    def __init__(self):
        self._dados = [None]*10
        self._indiceVaiSair = 0
        self._indiceEntrou = len(self._dados)-1
        self._tamanho = 0
    def __len__(self):
        return self._tamanho
    def _incrementa(self, indice):
        indice += 1
        if indice == len(self._dados):
            indice = 0
        return indice
    def _duplique_seq(self):
        novo = [None]*2*len(self._dados)
        for i in range(self._tamanho):
            novo[i] = self._dados[self._indiceVaiSair]
            self._indiceVaiSair = self._incrementa(self._indiceVaiSair)
            
        self._dados = novo
        self._indiceVaiSair = 0
        self._indiceEntrou = self._tamanho - 1
    def enqueue(self, x):
        if self._tamanho == len(self._dados):
            self._duplique_seq()
        self._indiceEntrou = self._incrementa(self._indiceEntrou)
        self._dados[self._indiceEntrou] = x
        self._tamanho += 1
    def dequeue(self):
        if self._tamanho == 0:
            raise IndexError("Fila vazia")
        self._tamanho -= 1
        x = self._dados[self._indiceVaiSair]
        self._dados[self._indiceVaiSair] = None
        self._indiceVaiSair = self._incrementa(self._indiceVaiSair)
        return x
    def isEmpty(self):
        return self._tamanho == 0
    def content(self):
        print(self._dados)

#Breadth-first search
def bfs(graph, start, goal):
    open_nodes = fila_circular()
    open_nodes.enqueue(start)
    closed_nodes = []
    steps = 0
    while not open_nodes.isEmpty():
        node = open_nodes.dequeue()
        if node == goal:
            return [True, steps]
        steps += 1
        for nextNode in graph[node]:
            if nextNode not in closed_nodes:
                open_nodes.enqueue(nextNode)
        closed_nodes.append(node)
    return [False, steps]

#Depth-first search
def dfs(graph, start, goal):
    open_nodes = [start]
    closed_nodes = []
    steps = 0
    while len(open_nodes) > 0:
        node = open_nodes.pop()
        if node == goal:
            return [True, steps]
        steps += 1
        for nextNode in graph[node]:
            if nextNode not in closed_nodes:
                open_nodes.append(nextNode)
        closed_nodes.append(node)
    return [False, steps]

#Iterative depth-first search
def DLS(graph, src, target, maxDepth, steps):
    if src == target: return [True, steps]
    if maxDepth <= 0: return [False, steps]
    steps += 1
    for i in graph[src]:
        r = DLS(graph, i, target, maxDepth-1, steps)
        steps = r[1]
        if(r[0]):
            return [True, steps]
    return [False, steps]

def IDDFS(graph, src, target, maxDepth):
    steps = 0
    for i in range(maxDepth):
        r = DLS(graph, src, target, i, steps)
        if(r[0]):
            return [True, r[1]]
        steps = r[1]
    return [False, steps]

# This class represents a directed graph using adjacency 
# list representation 
class Graph: 
  
    def __init__(self, g): 
        self.V = np.size(g, 0) # No. of vertices 
        self.E = 0             # No. of edges
        self._graphAM = g      # default adjacence matrix
        self._graphIL = None     # default dictionary to store graph's adjacence list
        if self._graphIL == None:
            self._adjMatrix2adjList()
    def _adjMatrix2adjList(self):
        v = 0
        self._graphIL = {}
        for line in range(np.size(self._graphAM, 0)):
            y = self._graphAM[line][:]
            try:
                self._graphIL[v] = [i for i, n in enumerate(y) if n > 0]
            except:
                self._graphIL[v] = ['']
            v += 1
    def getAM(self):
        return self._graphAM
    def getAL(self):
        return self._graphIL
    
    def _vertexMinDist(self, distSet, sptSet):
        minval = 99999
        for v in range(self.V):            
            if distSet[v] < minval and v not in sptSet:
                minval = distSet[v]
                minv = v
                
        return minv
    
    def trackPath(self, parentSet, j):  
        #Base Case : If j is source 
        track = [j]
        while parentSet[j] != -1:
            j = parentSet[j]
            track.append(j)
        return track
        
    
    def ucs(self, start, goal):
        sptSet = []       #Keeps track of vertices included in shortest path tree
        parentSet = [-1] * self.V
        distSet = [99999] * self.V
        distSet[start] = 0
        while goal not in sptSet and len(sptSet) < len(distSet):
            minv = self._vertexMinDist(distSet, sptSet)
            sptSet.append(minv)
            
            for vertex in self._graphIL[minv]:
                if distSet[minv] + self._graphAM[minv][vertex] < distSet[vertex]:
                    distSet[vertex] = distSet[minv] + self._graphAM[minv][vertex]
                    parentSet[vertex] = minv
        
        return([distSet[goal], self.trackPath(parentSet, goal)])

    
    
def main():
    g = np.loadtxt('valeAdjMatrix.txt')
    graph = Graph(g)
    #Nodos uteis: 
        # INF
        #   |
        #   |- parada INF: 22
        #   |- Predio 67
        #   |- RU de cima: 30
        #   |- Salas de aula INF: 59
        #   |- PontoCampus de cima: 29
        #   |- Topo escadaria Vale-INF: 58
        # VALE
        #   |
        #   |- PÃ© escadaria Vale-INF: 64
        #   |- RU de baixo: 185
        #   |- Parada Vale: 123
        #   |- Predio F: 181
        #   |- Predio do Calculo: 66
        
    start = 66
    goal = 185
    maxDepth = 12
    graph_al = graph.getAL()
    
    ucsl = graph.ucs(start, goal)
    print(ucsl[0], "meters")
    print("Path:",ucsl[1])
    bfsl = bfs(graph_al, start, goal)
    dfsl = dfs(graph_al, start, goal)
    iddfs = IDDFS(graph_al, start, goal, maxDepth)
    print("Busca em largura: ", bfsl)
    print("Busca em profundidade: ", dfsl)
    print("Busca em aprofundamento iterativo em",maxDepth, "niveis: ", iddfs)
    
if __name__ == "__main__":
    main()
```
