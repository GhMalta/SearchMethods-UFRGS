**Metodos de Busca Cega**

1) Busca em Largura



```python
import numpy as np

class fila_circular():
    def __init__(self):
        self._dados = [None]*10
        self._indiceVaiSair = 0
        self._indiceEntrou = len(self._dados)-1
        self._tamanho = 0
    def __len__(self):
        return self._tamanho
    def _incrementa(self, indice):
        indice += 1
        if indice == len(self._dados):
            indice = 0
        return indice
    def _duplique_seq(self):
        novo = [None]*2*len(self._dados)
        for i in range(self._tamanho):
            novo[i] = self._dados[self._indiceVaiSair]
            self._indiceVaiSair = self._incrementa(self._indiceVaiSair)
            
        self._dados = novo
        self._indiceVaiSair = 0
        self._indiceEntrou = self._tamanho - 1
    def enqueue(self, x):
        if self._tamanho == len(self._dados):
            self._duplique_seq()
        self._indiceEntrou = self._incrementa(self._indiceEntrou)
        self._dados[self._indiceEntrou] = x
        self._tamanho += 1
    def dequeue(self):
        if self._tamanho == 0:
            raise IndexError("Fila vazia")
        self._tamanho -= 1
        x = self._dados[self._indiceVaiSair]
        self._dados[self._indiceVaiSair] = None
        self._indiceVaiSair = self._incrementa(self._indiceVaiSair)
        return x
    def isEmpty(self):
        return self._tamanho == 0
    def content(self):
        print(self._dados)

#Breadth-first search
def bfs(graph, start, goal):
    open_nodes = fila_circular()
    open_nodes.enqueue(start)
    closed_nodes = []
    steps = 0
    while not open_nodes.isEmpty():
        node = open_nodes.dequeue()
        if node == goal:
            return [True, steps]
        steps += 1
        for nextNode in graph[node]:
            if nextNode not in closed_nodes:
                open_nodes.enqueue(nextNode)
        closed_nodes.append(node)
    return [False, steps]

#Depth-first search
def dfs(graph, start, goal):
    open_nodes = [start]
    closed_nodes = []
    steps = 0
    while len(open_nodes) > 0:
        node = open_nodes.pop()
        if node == goal:
            return [True, steps]
        steps += 1
        for nextNode in graph[node]:
            if nextNode not in closed_nodes:
                open_nodes.append(nextNode)
        closed_nodes.append(node)
    return [False, steps]

#Iterative depth-first search
def ids(graph, start, goal, maxDepth):
    steps = 0
    for currentMaxDepth in range(maxDepth):
        open_nodes = [start]
        depth_stack = [0]
        closed_nodes = []
        while len(open_nodes) > 0:
            node = open_nodes.pop()
            nodeDepth = depth_stack.pop()
            if node == goal:
                return [True, steps]
            steps += 1
            if nodeDepth < currentMaxDepth:
                for nextNode in graph[node]:
                    if (nextNode not in closed_nodes):
                        open_nodes.append(nextNode)
                        depth_stack.append(nodeDepth+1)                 
                closed_nodes.append(node)            
    return [False, steps]

def main():
    graph = {}
    g = open('graph1', 'r')
    for line in g:
        x, *y = line.split()
        graph[x] = y
    start = '0'
    bfsl = []
    dfsl = []
    idsl = []
    maxDepth = len(graph)
    for goal in range(maxDepth):
        goal = str(goal)
        bfsl.append(bfs(graph, start, goal)[1])
        dfsl.append(dfs(graph, start, goal)[1])
        idsl.append(ids(graph, start, goal, maxDepth)[1])
        #print("Breadth-first: ",bfs(graph, start, goal))
        #print("Depth-first: ", dfs(graph, start, goal))
        #print("IDS-first: ", ids(graph, start, goal, maxDepth))
    print("Media busca em largura: ", np.mean(bfsl))
    print(bfsl)
    print("Media busca em profundidade: ", np.mean(dfsl))
    print(dfsl)
    print("Media busca em aprofundamento iterativo: ", np.mean(idsl))
    print(idsl)
    
if __name__ == "__main__":
    main()
```
